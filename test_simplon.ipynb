```python
## Expliquer les détails affichés du résumé technique

Le résumé technique des données du fichier CSV fournit des informations importantes sur les données que nous avons chargées dans notre DataFrame.

A savoir :
    Le nombre de lignes indique combien de lignes de données sont présentes dans le fichier CSV. 
    Cela nous donne une idée de l'étendue de nos données.

    La liste des colonnes nous montre quelles catégories dinformations sont incluses dans le fichier CSV. 
    Ces colonnes sont les différents attributs ou caractéristiques de nos données.

    Les types de données des champs nous indiquent quel type de données est stocké dans chaque colonne.
    Par exemple, un champ peut contenir des nombres entiers (integers), des nombres décimaux (floats) ou des chaînes de caractères (strings).
    Cela est essentiel pour comprendre comment les données sont structurées et pour effectuer des analyses appropriées.
```

```python
#Calculer la médiane & moyen age

# med-age
med_age = df["Age"].median()

#moy-age
moy_age = df["Age"].mean()

print('la médiane de la colonne age est: ', med_age, 'et sa moyenne est: ', moy_age)
```

```python
#Calculer la médiane & moyen customer

#med_cust
med_cust = df["Customer spendings"].median()

#moy_cust

moy_cust = df["Customer spendings"].mean()

print('la médiane de la colonne age est: ', med_cust, 'et sa moyenne est: ', moy_cust)
```

```python
#Créer une visualisation des données du graphique à barres qui montre les dépenses des clients par pays

# Grouper les données par pays et calculer la somme des dépenses des clients pour chaque pays
pays_spendings = df.groupby('Country')['Customer spendings'].sum()

# création du graphe

plt.figure(figsize=(10, 6))
pays_spendings.sort_values(ascending=False).plot(kind='bar')
plt.title('Dépenses de clients par pays')
plt.xlabel('pays')
plt.ylabel('Dépenses des clients')
plt.xticks(rotation=45)
plt.tight_layout()
#Affichage du graphe
plt.show()
```

```python
# Nettoyer les lignes avec moins de 10 € de dépenses par client 

df = df[df['Customer spendings'] >= 10]
print(df.head())
```

```python
# Suppression des doublons

df_unique = df.drop_duplicates()

print(df_unique.head())
```

```python
# Nouveau df

df_cols =['Country', 'Age', 'Gender', 'Customer spendings']
df_finish = df_unique[df_cols]
df_finish.to_csv('fichier-final.csv', index=False)
```
